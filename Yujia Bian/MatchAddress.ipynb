```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df_sold = pd.read_excel('csv/nyc_sold_2022.xlsx')
df_sold.head()
```

```python
df_sold.count()
```

* ADDERSS has no null values. However, there is "N/A" in the text of the ADDRESS column. We decided to drop them.

```python
df_sold['ADDRESS'].isnull().sum()
```

```python
df_sold['ADDRESS'].str.contains('N/A', case=False).sum()
```

```python
mask = df_sold['ADDRESS'].str.contains('N/A', case=False)
df_sold.drop(df_sold[mask].index, inplace=True)
```

```python
df_sold['ADDRESS'].count()
```

```python
df_sold['Clean Address'] = df_sold['ADDRESS'].str.split(',').str[0].str.replace(r'\s*-\s*', '')
```

```python
df_sold[df_sold['ADDRESS'] == '94-30 59TH AVENUE, 4J']
```

```python
df_permit = pd.read_excel('csv/nyc_permit.xlsx')
df_permit.head()
```

```python
df_permit['Address'] = df_permit['House No'].astype(str) + ' ' + df_permit['Street Name'].astype(str)
```

```python
df_permit['Clean Address'] = df_permit['Address'].str.replace(r'\s*-\s*', '')
```

```python
df_permit.head()
```

```python
df_sold['Match'] = df_sold['Clean Address'].isin(df_permit['Clean Address']).astype(int)
```

```python
df_sold['Match'].sum()
```

```python
df_sold.head()
```

```python
df_permit_LL = df_permit[df_permit['Applicant Business Name'].str.contains('L&L', na=False)]
df_permit_LL
```

```python
df_sold.loc[df_sold['Clean Address'].isin(df_permit_LL['Clean Address']), 'Match'] = 2
```

```python
df_sold[df_sold['Match']==2]
```

```python
df_permit.to_csv('csv/permit.csv', index=False)
df_sold.to_csv('csv/sold.csv', index=False)
```
